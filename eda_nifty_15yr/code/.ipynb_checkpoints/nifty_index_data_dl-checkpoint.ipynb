```python
# use python 3.6

import  datetime
from datetime import date
import calendar

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# financial data
from nsepy import get_history
from nsepy.derivatives import get_expiry_date
```

```python
nifty = get_history(symbol = "NIFTY",
                   start = date(2005,1,1),
                   end = date(2021,3,31),
                   index = True)

print(nifty.shape)
nifty.tail()
```

```python
# add difference
nifty['daily_change'] = nifty['Close'].diff()
nifty.head()
```

```python
# change over 2 and 3 days
nifty['2day_change'] = nifty['daily_change'].rolling(2).sum()
nifty['3day_change'] = nifty['daily_change'].rolling(3).sum()
nifty.head()
```
